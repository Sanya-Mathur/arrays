```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data=pd.read_csv('/content/sample_data/california_housing_test.csv')
data.head()
# Box Plot: Median house value by housing median age
plt.figure(figsize=(10,6))
sns.boxplot(x='housing_median_age', y='median_house_value', data=data)
plt.xticks(rotation=45)
plt.title('Box Plot of Median House Value by Housing Median Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
data['housing_median_age'].value_counts().plot(kind='bar')
plt.title('Bar Chart of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.show()
```

```python
selected_ages = data['housing_median_age'].value_counts().index[:5]  # Select top 5 housing median ages
double_bar_data = data[data['housing_median_age'].isin(selected_ages)]
grouped = double_bar_data.groupby('housing_median_age')[['total_rooms', 'total_bedrooms']].sum()
grouped.plot(kind='bar', figsize=(10,6))
plt.title('Total Rooms vs Total Bedrooms by Housing Median Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python
age_groups = pd.cut(data['housing_median_age'], bins=5, labels=["Very New", "New", "Mid-Age", "Old", "Very Old"])
age_grouped = data.groupby(age_groups)['households'].sum()

plt.figure(figsize=(8,8))
age_grouped.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Households by Housing Median Age')
plt.ylabel('')  # Remove default label
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['median_income'], data['median_house_value'], alpha=0.5)
plt.title('Scatter Plot: Median Income vs Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```
